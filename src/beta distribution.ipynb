```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import math

# Define parameters for the Beta distribution
n = 100
alpha = 0.1
l = math.floor((n+1)*alpha)

alpha = 0.1
beta_ = l


def first(n, alpha):
    l = math.floor((n+1)*alpha)
    return n + 1 - l

def sec(n, alpha):
    return math.floor((n+1)*alpha)
    
    
# Generate a range of x values
x = np.linspace(0.8, 1., 1000)

# Calculate the probability density function (PDF) for the Beta distribution
pdf_1 = beta.pdf(x, first(50, alpha), sec(50, alpha))
pdf_2 = beta.pdf(x, first(100, alpha), sec(100, alpha))
pdf_3 = beta.pdf(x, first(500, alpha), sec(500, alpha))
pdf_4 = beta.pdf(x, first(1000, alpha), sec(1000, alpha))

# Plot the Beta distribution
plt.plot(x, pdf_1, label=f'n=50')
plt.plot(x, pdf_2, label=f'n=100')
plt.plot(x, pdf_3, label=f'n=500')
plt.plot(x, pdf_4, label=f'n=1000')
plt.title(f'Coverage distribution with alpha = {alpha}')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)

plt.savefig('coverage_distribution.png')

plt.show()
```
